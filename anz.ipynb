```python
print(df['merchant_code'].value_counts())
print(df['merchant_code'].unique())
```

```python
df.head()
```

```python
df.drop(columns='merchant_code',inplace=True)
df.isnull().sum()
```

```python
df['merchant_id'].fillna(str(df['merchant_id'].mode()),inplace=True)
df['merchant_suburb'].fillna(str(df['merchant_suburb'].mode()),inplace=True)
df['merchant_state'].fillna(str(df['merchant_state'].mode()),inplace=True)
df['merchant_long_lat'].fillna(str(df['merchant_long_lat'].mode()),inplace=True)
```

```python
df.isnull().sum()
```

```python
df[['long','lat']] = df.long_lat.str.split(expand=True)
df.drop(columns='long_lat',inplace=True)
```

```python
df[['merch_long','merch_lat']]=df.merchant_long_lat.str.split(" -",expand=True)
df.drop(columns='merchant_long_lat',inplace=True)
df.head()
```
